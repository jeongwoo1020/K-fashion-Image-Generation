```python
#---------------------------------------
# 아우터에서 ATTRIBUTES 통해서 텍스트 생성
#---------------------------------------
def outer(labeling):
    outer_data = labeling["아우터"][0]

    # 각 속성 조합
    color = outer_data.get("색상")
    sub_color = outer_data.get("서브색상")
    sleeve = outer_data.get("소매기장")
    print_ = ", ".join(outer_data.get("프린트", [])) if outer_data.get("프린트") else ""
    category = outer_data.get("카테고리", "") if outer_data.get("카테고리") else "상의"
    length = outer_data.get("기장")
    fit = outer_data.get("핏")
    material = ", ".join(outer_data.get("소재", [])) + " 소재" if outer_data.get("소재") else ""
    detail = ", ".join(outer_data.get("디테일", [])) + " 디테일" if outer_data.get("디테일") else ""
    neckline = outer_data.get("넥라인")
    collar = outer_data.get("옷깃")

    # 앞쪽 설명
    front = " ".join(filter(None, [color, f"/{sub_color}" if sub_color else None, print_, sleeve, category]))

    # 괄호 안 속성들을 리스트로 만들어서 유효한 값만 join
    inner_items = list(filter(None, [
        neckline,
        collar,
        f"{length}기장" if length else None,
        f"{fit}핏" if fit else None,
        material,
        detail
    ]))
    inner = ", ".join(inner_items)

    text = f"{front}({inner})"
    return text.strip()
```

```python
#---------------------------------------
# 원피스에서 ATTRIBUTES 통해서 텍스트 생성
#---------------------------------------
def onepiece(labeling):
    onepiece_data = labeling["원피스"][0]

    # 각 속성 조합
    color = onepiece_data.get("색상")
    sub_color = onepiece_data.get("서브색상")
    sleeve = onepiece_data.get("소매기장")
    print_ = ", ".join(onepiece_data.get("프린트", [])) if onepiece_data.get("프린트") else ""
    length = onepiece_data.get("기장")
    fit = onepiece_data.get("핏")
    material = ", ".join(onepiece_data.get("소재", [])) + " 소재" if onepiece_data.get("소재") else ""
    detail = ", ".join(onepiece_data.get("디테일", [])) + " 디테일" if onepiece_data.get("디테일") else ""
    neckline = onepiece_data.get("넥라인")

    # 앞쪽 설명: 색상, 서브색상, 프린트, 소매기장
    front = " ".join(filter(None, [
        color,
        f"/{sub_color}" if sub_color else None,
        print_,
        sleeve,
        "원피스"
    ]))

    # 괄호 안 속성들을 리스트로 만들어서 유효한 값만 join
    inner_items = list(filter(None, [
        f"{length}기장" if length else None,
        f"{fit}핏" if fit else None,
        neckline,
        material,
        detail
    ]))
    inner = ", ".join(inner_items)

    text = f"{front}({inner})"
    return text.strip()
```

```python
#---------------------------------------
# 상의(top)에서 ATTRIBUTES 통해서 텍스트 생성
#---------------------------------------
def top(labeling):
    top_data = labeling["상의"][0]

    # 각 속성 조합
    color = top_data.get("색상")
    sub_color = top_data.get("서브색상")
    sleeve = top_data.get("소매기장")
    print_ = ", ".join(top_data.get("프린트", [])) if top_data.get("프린트") else ""
    category = top_data.get("카테고리", "") if top_data.get("카테고리") else "상의"
    length = top_data.get("기장")
    fit = top_data.get("핏")
    material = ", ".join(top_data.get("소재", [])) + " 소재" if top_data.get("소재") else ""
    detail = ", ".join(top_data.get("디테일", [])) + " 디테일" if top_data.get("디테일") else ""
    neckline = top_data.get("넥라인")
    collar = top_data.get("옷깃")

    # 앞쪽 설명
    front = " ".join(filter(None, [color, f"/{sub_color}" if sub_color else None, print_, sleeve, category]))

    # 괄호 안 속성들을 리스트로 만들어서 유효한 값만 join
    inner_items = list(filter(None, [
        neckline,
        collar,
        f"{length}기장" if length else None,
        f"{fit}핏" if fit else None,
        material,
        detail
    ]))
    inner = ", ".join(inner_items)

    text = f"{front}({inner})"
    return text.strip()
```

```python
#---------------------------------------
# 하의에서 ATTRIBUTES 통해서 텍스트 생성
#---------------------------------------
def bottom(labeling):
    bottom_data = labeling["하의"][0]

    # 각 속성 조합
    color = bottom_data.get("색상")
    sub_color = bottom_data.get("서브색상")
    category = bottom_data.get("카테고리") if bottom_data.get("카테고리") else "하의"
    print_ = ", ".join(bottom_data.get("프린트", [])) if bottom_data.get("프린트") else ""
    length = bottom_data.get("기장")
    fit = bottom_data.get("핏")
    material = ", ".join(bottom_data.get("소재", [])) + " 소재" if bottom_data.get("소재") else ""
    detail = ", ".join(bottom_data.get("디테일", [])) + " 디테일" if bottom_data.get("디테일") else ""

    # 앞쪽 설명: 색상, 서브색상, 프린트, 카테고리
    front = " ".join(filter(None, [
        color,
        f"/{sub_color}" if sub_color else None,
        print_,
        category
    ]))

    # 괄호 안 속성들을 리스트로 만들어서 유효한 값만 join
    inner_items = list(filter(None, [
        f"{length}기장" if length else None,
        f"{fit}핏" if fit else None,
        material,
        detail
    ]))
    inner = ", ".join(inner_items)

    text = f"{front}({inner})"
    return text.strip()
```

```python
import os
import json
#---------------------------------------
# 폴더 위치에 맞게 dictionary 만들어주는 함수
#---------------------------------------
def collect_info_dict(json_data):
    labeling = json_data["데이터셋 정보"]["데이터셋 상세설명"]["라벨링"]
    info_dict = { "아우터" : 0, "원피스": 0, "상의" : 0, "하의" : 0,
                 "아우터_text" : "", "원피스_text" : "", "상의_text" : "", "하의_text" : ""}
    if labeling.get("원피스") and len(labeling["원피스"][0]) > 0:
        info_dict["원피스"] += 1
        info_dict["원피스_text"] = onepiece(labeling)
    if labeling.get("아우터") and len(labeling["아우터"][0]) > 0:
        info_dict["아우터"] += 1
        info_dict["아우터_text"] = outer(labeling)
    if labeling.get("상의") and len(labeling["상의"][0]) > 0:
        info_dict["상의"] += 1
        info_dict['상의_text'] = top(labeling)
    if labeling.get("하의") and len(labeling["하의"][0]) > 0:
        info_dict["하의"] += 1
        info_dict["하의_text"] = bottom(labeling)
    text = "아우터: {0} 원피스: {1} 상의: {2} 하의: {3} ".format(info_dict["아우터"], info_dict["원피스"], info_dict["상의"], info_dict["하의"])

    return info_dict

def make_info_dict(folder_dir):
  results = []

  for json_filename in os.listdir(folder_dir):
    if json_filename.endswith(".json"):
        json_path = os.path.join(folder_dir, json_filename)

        with open(json_path, "r", encoding="utf-8") as f:
              json_data = json.load(f)
        # 함수 호출하여 텍스트 생성
              information = collect_info_dict(json_data)
              result_text = "아우터: {0} 원피스: {1} 상의: {2} 하의: {3} ".format(information["아우터"], information["원피스"], information["상의"], information["하의"])
          # 필요하다면 리스트에 저장
        results.append({
              "filename": json_filename,
              "패션스타일": result_text,
              "패션_dict" : {"아우터" : information['아우터'], '원피스' : information['원피스'], '상의' : information['상의'], '하의' : information['하의']},
              "아우터_text": information["아우터_text"],
              "원피스_text": information["원피스_text"],
              "상의_text": information["상의_text"],
              "하의_text": information["하의_text"]
        })
  df = pd.DataFrame(results)
  return df
```

```python
#---------------------------------------
# 제거할 조건들 제거하는 함수
#---------------------------------------
def choice_data(df):
  condition = [
      "아우터: 0 원피스: 0 상의: 0 하의: 1 ",
      "아우터: 0 원피스: 0 상의: 1 하의: 0 ",
      "아우터: 1 원피스: 0 상의: 1 하의: 0 ",
      "아우터: 1 원피스: 0 상의: 0 하의: 0 "
  ]

  df_chosen = df[~df["패션스타일"].isin(condition)]
  df_chosen.reset_index(drop=True, inplace=True)
  return df_chosen
```

```python
#---------------------------------------
# 결과값: 정제된 데이터프레임
#---------------------------------------
df_chosen = choice_data(df)
df_chosen['패션스타일'].value_counts()
```

```python
#---------------------------------------
# 생성된 텍스트를 하나의 문장으로 합쳐주는 함수
#---------------------------------------
def concat_text(df_chosen):
  texts = []
  for i in range(len(df_chosen)):
    text = ""
    top_num = df_chosen['패션_dict'][i]['아우터'] + df_chosen['패션_dict'][i]['원피스'] + df_chosen['패션_dict'][i]['상의']
    bottom_num = df_chosen['패션_dict'][i]['하의']
    # 원피스만 착용
    if df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 1 and bottom_num == 0:
      text += df_chosen['원피스_text'][i] + "를 착용."
    # 원피스와 하의 착용
    elif df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 1 and bottom_num == 1:
      text += df_chosen['원피스_text'][i] + "와 함께, "
      text += df_chosen['하의_text'][i] + "를 매칭"
    # 원피스와 아우터 혹은 상의 총 두개 착용
    elif df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 2 and bottom_num == 0:
      text += df_chosen['원피스_text'][i] + "와 함께, "
      if df_chosen['패션_dict'][i]['아우터'] == 1:
        text += df_chosen['아우터_text'][i] + "를 착용."
      else:
        text += df_chosen['상의_text'][i] + "를 착용."
    # 원피스와 아우터 혹은 상의 총 두개 + 하의 착용
    elif df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 2 and bottom_num == 1:
      text += df_chosen['원피스_text'][i] + "와 함께, "
      if df_chosen['패션_dict'][i]['아우터'] == 1:
        text += df_chosen['아우터_text'][i] + "를 착용하고, "
        text += df_chosen['하의_text'][i] + "를 매칭."
      else:
        text += df_chosen['상의_text'][i] + "를 착용하고, "
        text += df_chosen['하의_text'][i] + "를 매칭."
    # 네가지 싹다 착용
    elif df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 3 and bottom_num == 1:
      text += df_chosen['원피스_text'][i] + "와 함께, "
      text += df_chosen['상의_text'][i] + "를 착용하고, "
      text += df_chosen['아우터_text'][i] + "를 걸쳐서, "
      text += df_chosen['하의_text'][i] + "를 매칭."
    # 위쪽 세가지 모두 착용, 하의 미착용
    elif df_chosen['패션_dict'][i]['원피스'] == 1 and top_num == 3 and bottom_num == 0:
      text += df_chosen['원피스_text'][i] + "와 함께, "
      text += df_chosen['아우터_text'][i] + "를 걸쳐고, "
      text += df_chosen['상의_text'][i] + "를 착용."
    # 아우터와 하의
    elif df_chosen['패션_dict'][i]['아우터'] == 1 and top_num == 1 and bottom_num == 1:
      text += df_chosen['아우터_text'][i] + "와 함께, "
      text += df_chosen['하의_text'][i] + "를 매칭."
    # 상의와 하의
    elif df_chosen['패션_dict'][i]['상의'] == 1 and top_num == 1 and bottom_num == 1:
      text += df_chosen['상의_text'][i] + "와 함께, "
      text += df_chosen['하의_text'][i] +"를 매칭."
    elif df_chosen['패션_dict'][i]['아우터'] == 1 and top_num == 2 and bottom_num == 1:
      text += df_chosen['상의_text'][i] + "와 함께, "
      text += df_chosen['아우터_text'][i] + "를 걸치고, "
      text += df_chosen['하의_text'][i] + "를 매칭."
    texts.append(text)
  df_chosen['full_text'] = texts
```

```python
#---------------------------------------
# 결과값: df_chosen의 full_text라는 새로운 칼럼
#---------------------------------------
concat_text(df_chosen)
```

```python
#---------------------------------------
# 전체 텍스트 기준 겹치는 값 제거하는 함수
#---------------------------------------
def remove_fulltext_duplicates(df_chosen):
    df_chosen.drop_duplicates(subset=['full_text'], keep='first', inplace=True)
    df_chosen.reset_index(drop=True, inplace=True)
    df_chosen['full_text'] = df_chosen['full_text'].str.replace(r'\(\)', '', regex=True)
    df_chosen.drop(columns=['패션_dict', '아우터_text', '원피스_text', "상의_text", "하의_text"], inplace=True)
    return df_chosen
```

```python
#---------------------------------------
# 결과값: 중복값 제거된 후, 필요없는 칼럼 제거한 데이터프레임
#---------------------------------------
remove_fulltext_duplicates(df_chosen)
```
